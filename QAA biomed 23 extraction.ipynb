```python
startpoint =38
section =0
subsection =0
subsubsection =0
text=''
subtext =''
points=[]
for p in doc.paragraphs[startpoint:]:
    if p.style.name != 'Normal' or not p.text:
        continue
    if p.text[0].isnumeric():
        if text:
            print('Adding point',p.text)
            points.append([section, subsection, subsubsection, text, subtext])
            subtext=''
            subsection=0
            subsubsection=0
        (designator, text )=p.text.split(' ',1)
        (section, subsection) = designator.split('.')
    
    else:
        print('new subsection',p.text)
        if subsubsection:
            points.append([section, subsection, subsubsection, text, subtext])
        subsubsection +=1
        subtext = p.text
        
```
